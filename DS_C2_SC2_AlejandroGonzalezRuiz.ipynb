a. Extraer la información del archivo.

```python
vacunas = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Modulo de aplicacion Pandas/country_vaccinations.csv')
vacunas.head(5) # visualizamos los primeros 5registros
```

b. Mostrar la estructura y tipos de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas, asegurándote que las columnas con fechas sean del tipo datetime64

```python
vacunas.info() # visualizamos los tipos de datos, para identificar las operaciones que podremos hacer
vacunas['date'] = pd.to_datetime(vacunas['date']) ## Al no ser un tipo de datos datetime, usamos este comando para cambiar el Dtype
vacunas.info() # validamos nuevamente
```

c. Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines, en otras palabras, agrupe por vaccines y realice la sumatoria).

```python
newDf_vaccines =  vacunas.copy() #Guardamos una copia, para poder hacer las operaciones y manipular resultados parciales
newDf_vaccines['AmountVaccines'] = newDf_vaccines['vaccines'].str.split(', ').str.len() #   Creamos una columna derivada para poder almacenar las la cantidad de bacunas, la funcion len()
                                                                                        #   nos retorna el valor en enteros
newDf_vaccines['AmountVaccines'] #  visualizamos la columna derivada con la información requerida.
```

d. Obtener la cantidad de vacunas aplicadas en todo el mundo

```python
# Usamos la función sum() y tomamos el campo "daily_vaccinations" para obtener el total de vacunas según el Data Frame
print("El total de vacunas aplicadas en el mundo es:", vacunas['daily_vaccinations'].sum())
```

e. Calcular el promedio de vacunas aplicadas por país

```python
# usamos la función mean() para el promedio y la función goupby() para la discriminación de columnas
vacunas[['country','daily_vaccinations']].groupby('country').mean()
```

f. Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo

```python

vacunas01292021 = vacunas[vacunas['date'] == '29-01-2021']  # Guardamos en un nuevo DataFrame con las lineas que tengan ese valor en fecha
print('la cantidad de vacunas aplicadas el día 29/01/2023 es:', vacunas01292021['daily_vaccinations'].sum()) #Tomamos la columna daily_vaccinations pues es el campo que necesitamos para obtener le numero de vacunas

```

g. Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  

```python
conDiferencias = vacunas.copy() #creamos la copia del DataFrame
conDiferencias['diferencias'] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']  # hacemos la diferencia
conDiferencias # traemos el datafra
```

h. Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
# utilizamos las funciones .max() para la fecha mas reciente .min() para la fecha más antigua
periodo = vacunas['date'].max() - vacunas['date'].min()
print(' el periodo de timepo entre los registros es de: ',periodo)
```

i. Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter "," ).

```python
conCantidad = vacunas.copy()
conCantidad['conVac'] = conCantidad['vaccines'].str.split(',').str.len()
conCantidad
```

j. Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = vacunas[vacunas['date'] < '20-12-2020']  # utilizamos unicamente < porque el enunciado dice "antes"
antes20
```

k. Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = vacunas[vacunas['vaccines'].str.contains('Pfizer')]  # utilizamos unicamente < porque el enunciado dice "antes"
pfizer
```
